###  1. Write a python program to create an array and covert to a desired dimension, as shown in picture.

```python
list1 = [[8,9,10],
        [11,12,13]]
import numpy as np
array_1 = np.array(list1)
array_2 = array_1.reshape(3,2)
print(array_1)
print("Reshaped array")
print(array_2)
```

### 4.Read the given data set and answer the following questions. 

```python
# 1.Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.
diab_data = pd.read_csv("diabetes_model.csv")
indexed_df = diab_data[["Glucose","Age"]].iloc[45:61]
indexed_df
```

```python
# 2. How many patients in the data are having the blood pressure level above 120?
diab_data[diab_data["BloodPressure"] > 120]
```

###  Only 1 person is having the blood pressure level above 120.

```python
# 3. Select the rows where the age is between 24 and 29 (inclusive)
diab_data.loc[(diab_data['Age']>=24) & (diab_data['Age']<=29)]
```

```python
# 4. Find the minimum and maximum value of insulin level in the data.
diab_data["Insulin"].max()
```

```python
diab_data["Insulin"].min()
```

```python
# 5. Create a histogram of "Age".
import matplotlib.pyplot as plt
diab_data.hist(column='Age',figsize=(8,8),color = 'pink',legend=True)
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python
# 6. Create a scatter plot of "BMI" vs "Glucose".
import seaborn as sns
sns.scatterplot(diab_data['BMI'], diab_data['Glucose'])

plt.title('BMI v/s Glucose Plot')
plt.xlabel("BMI")
plt.ylabel("Glucose")
```

```python
# 7. Replace the min value of "BMI" with mean value of BMI
diab_data['BMI'].min()
```

```python
diab_data['BMI'].mean()
```

```python
diab_data['BMI'].replace(diab_data['BMI'].min(), diab_data['BMI'].mean(), inplace = True)
```

```python
diab_data['BMI'].min()
```

```python
# 8. Draw a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome.
plt.figure(figsize=(8,8))
sns.boxplot(x="Outcome",y="Glucose",data=diab_data)
plt.title("Boxplot of Outcome vs Glucose ")
```
